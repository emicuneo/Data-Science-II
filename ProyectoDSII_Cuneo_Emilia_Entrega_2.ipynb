### TRATAMIENTO DE DATOS NULOS

```python
# Se detectan y trabaja con los datos nulos. Para eso se usa una funcion que arma una tabla con las variables con 
# datos faltantes y cantidades y tambien con sus porcentajes para ver que tan representativos son
def get_tabla_valores_faltantes(df):
    mis_val = df.isnull().sum()
    mis_val_percent = 100 * df.isnull().sum()/len(df)
    mis_val_table = pd.concat([mis_val,mis_val_percent], axis = 1)
    mis_val_table_columns = mis_val_table.rename(columns = {0: 'VALORES FALTANTES', 1: '% DEL TOTAL DE VALORES'})
    mis_val_table_columns = mis_val_table_columns[mis_val_table_columns.iloc[:,1] != 0]
    mis_val_table_columns = mis_val_table_columns.sort_values('% DEL TOTAL DE VALORES', ascending = False).round(1)
    print('El data frame tiene ' + str(df.shape[1]) + ' columnas.\n' 'Hay ' + str(mis_val_table_columns.shape[0]) + 
          ' columnas con valores faltantes')
    return mis_val_table_columns
    
```

```python
# Se pasa como argumento de la funcion el data frame que estamos usando
get_tabla_valores_faltantes(df_LED)
```

Se puede ver que hay un número considerable de variables con datos faltantes, y algunas representan cerca del 20 % de los datos. No vamos a eliminar ninguna columna porque son importantes para el análisis.Por esto, se imputarán los nulos para poder trabajar siguiendo distintos criterios según el caso.  
. Para el caso de la variable **Population** se podría recurrir a fuentes externas para rellenar los valores, teniendo en cuenta que es información que puede ser accesible. En este caso, como es un trabajo de clase y teniendo en cuenta el tiempo que llevaría buscar cada valor, se realizará una imputación con interpolación para cada país.  
. Para el caso de la variable **Hepatitis B** se realizará una imputación con la mediana por status, es decir dependiendo de si el pais esta en desarrollo o no, teniendo en cuenta que puede variar mucho el valor dependiendo de esto para el estado de vacunación.  
. Para el caso de la variable **GDP** se realizará una imputación con interpolación para cada país.  
. Para la variable **Total expenditure** se realizará una imputación con interpolación por cada país.  
. Para la variable **Alcohol** se realizará una imputación con la mediana por cada país.  
. Para la variable **Income composition of resources** se realizará una imputación con interpolación por cada país.  
. Para la variable **Schooling** se realizará una imputación con interpolación por cada país.  
. Para la variable **BMI** se realizará una imputación con la mediana por status.  
. Para las variables **Thinness** se realizará una imputación con la mediana por status.  
. Para la variable **Polio** se realizará una imputación con la mediana por cada status.  
. Para la variable **Diphtheria** se realizará una imputación con la mediana por status.  
. Para la variable **Life Expectancy** se realizará una imputación con interpolación por país.  
. Para la variable **Adult Mortality** se realizará una imputación con interpolación por país.  


```python
# Imputación de variables
#Population
df_LED['Population'] = df_LED.groupby('Country', group_keys = False)['Population'].apply(lambda x: x.interpolate())
df_LED['Population'] = df_LED['Population'].fillna(method='bfill').fillna(method = 'ffill')  
#Hepatitis B
df_LED['Hepatitis B'] = df_LED.groupby('Status')['Hepatitis B'].transform(lambda x: x.fillna(x.median()))  
#GDP
df_LED['GDP'] = df_LED.groupby('Country', group_keys = False)['GDP'].apply(lambda x: x.interpolate())
df_LED['GDP'] = df_LED['GDP'].fillna(method='bfill').fillna(method = 'ffill')  
#Total expenditure
df_LED['Total expenditure'] = df_LED.groupby('Country', group_keys = False)['Total expenditure'].apply(lambda x: x.interpolate())
df_LED['Total expenditure'] = df_LED['Total expenditure'].fillna(method='bfill').fillna(method = 'ffill')  
#Alcohol
df_LED['Alcohol'] = df_LED.groupby('Status')['Alcohol'].transform(lambda x: x.fillna(x.median()))  
#Income composition
df_LED['Income composition of resources'] = df_LED.groupby('Country', group_keys = False)['Income composition of resources'].apply(lambda x: x.interpolate())
df_LED['Income composition of resources'] = df_LED['Income composition of resources'].fillna(method='bfill').fillna(method = 'ffill') 
#Schooling
df_LED['Schooling'] = df_LED.groupby('Country', group_keys = False)['Schooling'].apply(lambda x: x.interpolate())
df_LED['Schooling'] = df_LED['Schooling'].fillna(method='bfill').fillna(method = 'ffill') 
#BMI
df_LED[' BMI '] = df_LED.groupby('Status')[' BMI '].transform(lambda x: x.fillna(x.median()))
#Thiness
df_LED[' thinness  1-19 years'] = df_LED.groupby('Status')[' thinness  1-19 years'].transform(lambda x: x.fillna(x.median()))
df_LED[' thinness 5-9 years'] = df_LED.groupby('Status')[' thinness 5-9 years'].transform(lambda x: x.fillna(x.median()))
#Polio
df_LED['Polio'] = df_LED.groupby('Status')['Polio'].transform(lambda x: x.fillna(x.median()))
#Diphtheria
df_LED['Diphtheria '] = df_LED.groupby('Status')['Diphtheria '].transform(lambda x: x.fillna(x.median()))
#Life Expectancy
df_LED['Life expectancy '] = df_LED.groupby('Country', group_keys = False)['Life expectancy '].apply(lambda x: x.interpolate())
df_LED['Life expectancy '] = df_LED['Life expectancy '].fillna(method='bfill').fillna(method = 'ffill') 
#Adult Mortality
df_LED['Adult Mortality'] = df_LED.groupby('Country', group_keys = False)['Adult Mortality'].apply(lambda x: x.interpolate())
df_LED['Adult Mortality'] = df_LED['Adult Mortality'].fillna(method='bfill').fillna(method = 'ffill') 
```

### MODELO DE MACHINE LEARNING  

Luego de haber realizado la limpieza, tratamiento de los valores nulos y exploración y visualización de datos, se prosigue con la definición del modelo de machine learning que se va a utilizar para la finalidad de este trabajo. En este caso, que se están estudiando los factores que afectan la esperanza de vida de los países, lo que queremos predecir es justamente la esperanza de vida de un país. 

Primero, definimos que se utilizará un algoritmo supervisado. Esto es así ya que contamos con una columna de etiqueta o label, que nos permite usar de referencia para crear el modelo. En este caso, este dato es la columna esperanza de vida. Dentro de los algorítmos supervisados, se selecciona el modelo de regresión lineal, ya que lo que buscamos es predecir una variable numérica, llamada variable dependiente(esperanza de vida) en función de una variable o variables independientes(todas las demas columnas).

```python
# One-Hot Encoding para la variable Status
df_LED = pd.get_dummies(df_LED, columns=['Status'], drop_first = True)
```

Se transforma las variables categóricas en numéricas para que el modelo pueda procesarlas. Para el caso de la variable Status se usa One Hot Encoding porque solo tiene dos opciones: en desarrollo y desarrollado y Label Encoding para los países ya que son muchos más valores.

```python
# Se modifica de valores booleanos( True- False a 0 y 1 para que el modelo funcione)
df_LED['Status_Developing'] = df_LED['Status_Developing'].astype(int)
```

```python
#Se separa el target de las demas variables
x = df_LED.drop(columns=['Life expectancy '])
y = df_LED['Life expectancy ']
```

Se prueba en primera instancia un modelo de Regresión Lineal

```python
# Modelo de regresión lineal
model = LinearRegression()
```

```python
# Se utilizá K-Fold cross-validation para separar en train y test, con k = 10
kf = KFold(n_splits = 10, shuffle = True, random_state = 42)
```

```python
# Se entrena el modelo y se calculan las métricas en un for para k = 10

rmse_scores = []
r2_scores = []
mae_scores = []

for train_index, test_index in kf.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    rmse = np.sqrt(mean_squared_error(y_test,y_pred))
    r2 = r2_score(y_test,y_pred)
    mae = mean_absolute_error(y_test,y_pred)
    print(rmse,r2,mae)
    rmse_scores.append(rmse)
    r2_scores.append(r2)
    mae_scores.append(mae)
    
    
```

```python
# Se realizan los promedios de todas las metricas a partir de los 10 modelos
print(f'rmse promedio: {np.mean(rmse_scores):.2f}')
print(f'r2 promedio: {np.mean(r2_scores):.2f}')
print(f'mae promedio: {np.mean(mae):.2f}')
```

Que el valor de r2 haya dado 0.81 significa que el modelo explica el 81 % de la variabilidad de la esperanza de vida. Pero todavia hay un 19 % de variabilidad que no se puede explicar con este modelo.
El MAE es el promedio de las diferencias absolutas entre los valores predichos y los reales.
El RMSE es la raiz cuadrada del promedio de los errores al cuadrado. Penaliza mas los errores grandes.
Por lo tanto se puede ver que el modelo se equivoca, en promedio, en 4,16 años( para arriba o para abajo) al predecir la esperanza de vida.
No parece un tan mal modelo, pero vamos a probar usar otro modelo y tratar de mejorarlo también.

Se procede a hacer una modificación del hyperparametro k para ver si el modelo mejora al probar otras alternativas

```python
# Se asignan distintos valores a k con los que se probaran los modelos
k_values = [5, 8, 10, 12, 15]
```

```python
# Se entrena el modelo y se calculan las métricas en un for para los distintos valores de k
resultados = {
    'k': [],
    'rmse': [],
    'r2': [],
    'mae': []
}

rmse_scores = []
r2_scores = []
mae_scores = []

for k in k_values:
    kf = KFold(n_splits = k, shuffle = True, random_state = 50)
    for train_index, test_index in kf.split(x):
        x_train, x_test = x.iloc[train_index], x.iloc[test_index]
        y_train, y_test = y.iloc[train_index], y.iloc[test_index]
        model.fit(x_train,y_train)
        y_pred = model.predict(x_test)
        rmse = np.sqrt(mean_squared_error(y_test,y_pred))
        r2 = r2_score(y_test,y_pred)
        mae = mean_absolute_error(y_test,y_pred)
        rmse_scores.append(rmse)
        r2_scores.append(r2)
        mae_scores.append(mae)
    resultados['k'].append(k)
    resultados['rmse'].append(np.mean(rmse_scores))
    resultados['r2'].append(np.mean(r2_scores))
    resultados['mae'].append(np.mean(mae_scores))
    print(f'Para k = {k}: rmse promedio = {np.mean(rmse_scores)}, r2 promedio = {np.mean(r2_scores)}, mae promedio = {np.mean(mae_scores)}')
```

Se puede ver que los valores de rmse, r2 y mae no difieren mucho a medida que se prueban distintos valores de k. Es decir que modificar el valor de k no produce un impacto significativo. Por lo tanto para este caso la mejor opción es usar k = 5, que tiene buenos resultados y es más eficiente por el tiempo de procesamiento.

Como siguiente paso, se probará usar otro modelo para el problema. Se utilizará XGBoost y también se aplicará, como en el caso anterior K-Fold Cross Validation para divir en k cantidad de grupos, donde k-1 son los que se utilizarán para entrenar al modelo y uno para la validación.


```python
# Se inicializa el modelo XGBoost
xgb_model = xgb.XGBRegressor(objective = 'reg:squarederror', n_estimators = 100, learning_rate = 0.1, max_depth = 5, random_state = 42)
```

```python
# K-Fold Cross Validation
kf = KFold(n_splits = 5, shuffle = True, random_state = 42)
```

```python
# Se entrena el modelo y se calculan las métricas en un for para un valor de k = 5
r2_scores = []
rmse_scores = []

for train_index, test_index in kf.split(x):
    x_train, x_test = x.iloc[train_index], x.iloc[test_index]
    y_train, y_test = y.iloc[train_index], y.iloc[test_index]
    xgb_model.fit(x_train, y_train)
    y_pred = xgb_model.predict(x_test)
    r2 = r2_score(y_test, y_pred)
    rmse = np.sqrt(mean_squared_error(y_test, y_pred))
    r2_scores.append(r2)
    rmse_scores.append(rmse)
    
print(f"R² promedio: {np.mean(r2_scores):.4f}")
print(f"RMSE promedio: {np.mean(rmse_scores):.4f}")
```

```python
# Definir el modelo base XGBoost
xgb_model = XGBRegressor(objective='reg:squarederror', random_state=42)

# Definir el rango de valores para los hiperparámetros
param_grid = {
    'n_estimators': [100, 200],  # Número de árboles
    'learning_rate': [0.01, 0.1, 0.2],  # Tasa de aprendizaje
    'max_depth': [3, 5, 7],  # Profundidad máxima de los árboles
    'min_child_weight': [1, 3, 5],  # Peso mínimo de las hojas
    'subsample': [0.7, 0.8, 1.0],  # Fracción de datos para entrenar
    'colsample_bytree': [0.7, 0.8, 1.0]  # Fracción de características
}

# Crear el objeto GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, 
                           scoring='neg_mean_squared_error', cv=5, n_jobs=-1, verbose=2)

# Ajustar el modelo con GridSearchCV
grid_search.fit(x, y)

# Mostrar los mejores parámetros
print(f"Mejores parámetros encontrados: {grid_search.best_params_}")

# Evaluar el modelo con los mejores parámetros
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)

# Calcular R² y RMSE con el mejor modelo
from sklearn.metrics import r2_score, mean_squared_error
r2 = r2_score(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)

print(f"R² con los mejores parámetros: {r2:.4f}")
print(f"RMSE con los mejores parámetros: {rmse:.4f}")
```

Con este método se encuentran los mejores parámetros para el modelo. Se puede ver que los valores de r2 y rmse mejoraron considerablemente en relación al
modelo anterior de regresión lineal y también a la primera prueba usando este modelo pero sin optimizar hiperparámetros.
